````python
df = pd.read_csv('data/NBA_train.csv')
df.head().T
````

````python
df['points_diff'] = df['PTS'] - df['oppPTS']

independent_vars = df['points_diff']
dependent_var = df['W']
sns.regplot(data=df ,x=independent_vars,y=dependent_var)
````

````python

X=sm.add_constant(independent_vars)

model_wins=sm.OLS(dependent_var,X).fit()
y_predit=model_wins.predict(X)

result=model_wins.summary()
print(result)
````

````python
mean_sq_error=rmse(dependent_var,y_predit)
mean_sq_error
````

````python
print('Dependent Variable %s average' % 'w',dependent_var.mean())
````

````python
# Extract wins for teams that made and didn't make the playoffs
playoff_wins = df[df['Playoffs'] == 1]['W'].values
non_playoff_wins = df[df['Playoffs'] == 0]['W'].values


from scipy.stats import norm

# Calculate mean and standard deviation for non-playoff wins
mean_non_playoff_wins = np.mean(non_playoff_wins)
std_non_playoff_wins = np.std(non_playoff_wins)

# Determine the wins above which 95% of the non-playoff teams fall
threshold_wins_95 = norm.ppf(0.95, mean_non_playoff_wins, std_non_playoff_wins)
threshold_wins_95
````

````python
all_numeric_vars = list(df.select_dtypes (include='number').columns)

print(all_numeric_vars)

dependent_variable = 'PTS'

for col in all_numeric_vars:
    try:
        sns. jointplot (x=col, y=dependent_variable, data=df, kind="reg");
    except:
        print('Could not plot variable %s' % col)
````

````python
independent_variables = ['2PA', '3PA', 'FTA']

X = df [independent_variables]

y = df['PTS']

X = sm.add_constant(X)

print (y.mean())

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

Ir_pts = LinearRegression()

Ir_pts.fit(X_train, y_train)

y_pred = Ir_pts.predict(X_test)

r_squared = metrics.r2_score (y_test, y_pred)

print('R-Squared Score:', r_squared)

mse_Ir_pts = metrics.mean_squared_error (y_test, y_pred)

rmse_Ir_pts = metrics.mean_squared_error (y_test, y_pred, squared=False)

np.sqrt(mse_Ir_pts), rmse_Ir_pts


model_runs_scored = sm.OLS(y, X).fit()
#predictions = mode_runs_scored.predict(X) 
print(model_runs_scored.summary())
````

````python
for col in independent_variables:
    sns.jointplot(x=col, y=y, data=df, kind="reg");
````

````python

plt.scatter(model_runs_scored.fittedvalues, model_runs_scored.resid, alpha=0.5)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.axhline(y = 0, color = 'r')
plt.show()
````
