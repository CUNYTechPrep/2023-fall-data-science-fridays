```python
# Importing necessary libraries for data analysis, visualization, and machine learning

import pandas as pd  # Pandas is used for data manipulation and analysis using dataframes
import numpy as np   # NumPy is a library for numerical operations, especially with arrays
import matplotlib.pyplot as plt  # Matplotlib is a plotting library for creating static, animated, and interactive visualizations
import seaborn as sns  # Seaborn is a statistical data visualization library based on Matplotlib, providing a high-level interface for drawing attractive and informative statistical graphics
######## scikit-learn modules ########
from sklearn.model_selection import train_test_split  # scikit-learn library for machine learning; train_test_split is used to split the dataset into training and testing sets
from sklearn.linear_model import LinearRegression  # LinearRegression is a class for linear regression models in scikit-learn
from sklearn.preprocessing import StandardScaler  # StandardScaler is used for standardizing features by removing the mean and scaling to unit variance
from sklearn.ensemble import RandomForestRegressor  # RandomForestRegressor is an ensemble model for regression tasks in scikit-learn, combining multiple decision trees for improved performance
```

```python
# Reading a CSV file named 'housing.csv' into a Pandas DataFrame
df = pd.read_csv('../data/housing.csv')
```

```python
df
```

```python
df.info()
```

```python
# The inplace=True parameter modifies the DataFrame in place, meaning it will be updated without the need for assignment.
# After this operation, the DataFrame will no longer contain rows with any missing values.

df.dropna(inplace=True)
```

```python
df.info()
```

```python
# Extracting features (independent variables) by dropping the 'median_house_value' column
X = df.drop(['median_house_value'], axis=1)

# Extracting the target variable (dependent variable) 'median_house_value'
y = df['median_house_value']
```

```python
# Split the dataset into training and testing sets
# X_train: Training set for features
# X_test: Testing set for features
# y_train: Training set for the target variable
# y_test: Testing set for the target variable
# The test_size parameter specifies the proportion of the dataset to include in the test split (here, 20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# The resulting DataFrame, train_data, will have the columns of X_train
# along with an additional column for y_train.

train_data = X_train.join(y_train)
```

```python
train_data
```

```python
#'hist' method to create a histogram, 'figsize' is a parameter specifying the dimensions of the figure (width, height).
train_data.hist(figsize=(15, 8))
```

```python
plt.figure(figsize=(15,8))

# train_data.corr(): Calculates the correlation matrix for the numerical columns in 'train_data'
# annot=True: Annotates each cell in the heatmap with the corresponding correlation coefficient
# cmap="YlGnBu": Specifies the color map to be used for the heatmap (Yellow-Green-Blue)

# The resulting heatmap provides a visual representation of how strongly each variable is correlated with every other variable
# Positive correlations are indicated by warmer colors (closer to blue), while negative correlations are indicated by cooler colors (closer to yellow)
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
# Logarithmic transformation of 'total_rooms' column to address skewness
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)

# Logarithmic transformation of 'total_bedrooms' column to address skewness
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)

# Logarithmic transformation of 'population' column to address skewness
train_data['population'] = np.log(train_data['population'] + 1)

# Logarithmic transformation of 'households' column to address skewness
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
# The 'value_counts()' method is used on the 'ocean_proximity' column to count
# the occurrences of each unique value in the column and return a Series with the counts.
# The assumption based on the heatmap is that the closer the homes are to the water, the more expensive they are.

# This line is helpful for understanding the distribution of different categories
# within the 'ocean_proximity' column of the training data.
train_data.ocean_proximity.value_counts()
```

```python
# DataFrame with one-hot encoded columns for 'ocean_proximity'.
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
plt.figure(figsize=(15,8))

# x-axis represents the "latitude" column from the DataFrame "train_data"
# y-axis represents the "longitude" column from the DataFrame "train_data"
# Points on the plot are colored based on the "median_house_value" column from the DataFrame "train_data"
# The color palette "coolwarm" is used, which transitions from cool to warm colors
sns.scatterplot(x="latitude", y="longitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
# Calculate the bedroom ratio by dividing the total number of bedrooms by the total number of rooms.
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']

# Calculate the average number of rooms per household by dividing the total number of rooms by the number of households.
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
# Extracting features (independent variables) and the target variable from the training data
X_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']

# Creating an instance of the LinearRegression model
reg = LinearRegression()

# Fitting the linear regression model using the training data
reg.fit(X_train, y_train)
```

```python
# Combine the feature variables (X_test) and the target variable (y_test)
test_data = X_test.join(y_test)

# Apply logarithmic transformation to certain features to handle skewed distributions
test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

# One-hot encode the categorical variable 'ocean_proximity' and drop the original column
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

# Calculate the bedroom ratio by dividing the total number of bedrooms by the total number of rooms
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']

# Calculate the average number of rooms per household by dividing the total number of rooms by the number of households
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
X_test, y_test = test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
```

```python
# Return the coefficient of determination of the prediction.
reg.score(X_test, y_test)
```

```python
# Create an instance of the StandardScaler
scaler = StandardScaler()

# Apply the StandardScaler to the training features (X_train)
# This process scales the features to have a mean of 0 and a standard deviation of 1
X_train_s = scaler.fit_transform(X_train)

# The scaler is now fitted to the training data, and its transformation parameters are stored
# These parameters (mean and standard deviation) will be used to transform any future data

# Fit the linear regression model (reg) to the scaled training data (X_train_s) and the target variable (y_train)
# The linear regression model learns the relationship between the scaled features and the target variable
reg.fit(X_train_s, y_train)
```

```python
# 'scaler.transform()' applies the same scaling transformation that was applied to the training set features (X_train_s) to the test set features. This ensures consistency in scaling between the training and test sets.
# (e.g., mean subtraction and division by standard deviation)

X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s, y_test)
```

```python
# Creating an instance of the RandomForestRegressor
forest = RandomForestRegressor()

# Training the RandomForestRegressor model on the training data
# X_train_s is the feature matrix (input data) for the training set
# y_train is the target variable for the training set
forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```
